```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['VisitorType'] = le.fit_transform(df['VisitorType'])
df['Month'] = le.fit_transform(df['Month'])
df['Weekend'] = le.fit_transform(df['Weekend'])
df['Revenue'] = le.fit_transform(df['Revenue'])

X1 = Df[['ProductRel_per_dur','Admin_per_dur','Inform_per_dur','PageValues','Region','Month']]
# X1 = Df.drop('Revenue',axis=1)
y1 = Df['Revenue']

from imblearn.combine import SMOTETomek
smk = SMOTETomek(random_state=42)
X_res,y_res = smk.fit_resample(X1,y1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2 

X1 = Df.drop('Revenue',axis=1)
y1 = df['Revenue']

#use inbuilt class feature_importances of tree based classifiers
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier(n_estimators = 5, 
                                        criterion ='entropy', max_features = 12)
model.fit(X1,y1)
plt.figure()

feat_importances = pd.Series(model.feature_importances_, index=X1.columns)
ax = feat_importances.nlargest(15).plot(kind='barh',color='grey')
plt.xlabel('Feature Importance Score')
plt.subplots_adjust(left=0.25)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

X1 = Df.drop('Revenue',axis=1)
y1 = Df['Revenue']

bestfeatures = SelectKBest(score_func=chi2, k=12)
fit = bestfeatures.fit(X1,df['Revenue'])

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X1.columns)

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']

print(featureScores.nlargest(12,'Score'))
ax = featureScores.nlargest(10,'Score').plot(kind='bar')
plt.subplots_adjust(left=0.25)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_res)
X1 = scaler.transform(X_res)

from sklearn.model_selection import train_test_split
X1,x2,y1,Y2 = train_test_split(X1,y_res,test_size=0.25)

from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
train = []
val = []
train_f1_score = []
val_f1_score = []
for i in range(1,21):
    neigh = KNeighborsClassifier(n_neighbors=i,algorithm='kd_tree',weights='distance')
    neigh.fit(X1, y1)
    train.append(neigh.score(X1,y1))
    val.append(neigh.score(x2,Y2))
    train_f1_score.append(f1_score(y1,neigh.predict(X1)))
    val_f1_score.append(f1_score(Y2,neigh.predict(x2)))
```

```python
plt.figure()
plt.title('Accuracy Metric')
plt.ylabel('Acc')
plt.xlabel('Number of K neigbours')
plt.plot(train)
plt.plot(val)
plt.figure()
plt.title('F1 Score Metric')
plt.ylabel('f1 score')
plt.xlabel('Number of K neigbours')
plt.plot(train_f1_score)
plt.plot(val_f1_score)
```
